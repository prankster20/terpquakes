```python
# Drop duplicates
before=len(df.index)
print(f"BEFORE: {before}")
df.drop_duplicates(inplace=True)
after=len(df.index)
print(f"AFTER: {after}")
print(f"REMOVED: {before-after}")


# print(df[((df.gap > 180) | (df.dmin >= 10)) & (df.mag < 5.5)].drop_duplicates().id.count())
# gap > 180 or dmin >= 10 with a mag < 5.5

# df.drop()
```

```python
# Drop rows with mag lower than 4
before=len(df.index)
print(f"BEFORE: {before}")
df.drop(df[df.mag<=4].index, inplace=True)
after=len(df.index)
print(f"AFTER: {after}")
print(f"REMOVED: {before-after}")
```

```python
# Replace 0 or NA nst values with 1
print("BEFORE")
print(df.nst[df.nst == 0].count())
print(df.nst[df.nst.isnull()].count())

# change 0 to 1
df.nst.replace(0,1, inplace=True)

print("AFTER")
print(df.nst[df.nst == 0].count())
print(df.nst[df.nst.isnull()].count())
```

```python
# Replace NA in magerror with 0
print(f"BEFORE (magError.isnull()): {len(df.magError[df.magError.isnull()])}")
df.magError.replace(NA,0,inplace=True)
print(f"AFTER  (magError.isnull()): {len(df.magError[df.magError.isnull()])}")

print(f"BEFORE (magNst.isnull() or 0): {len(df.magNst[(df.magNst.isnull()) | (df.magNst == 0)])}")
df.magNst.replace(NA,1,inplace=True)
df.magNst.replace(0,1,inplace=True)
print(f"AFTER  (magNst.isnull() or 0): {len(df.magNst[(df.magNst.isnull()) | (df.magNst == 0)])}")
```

```python
print(f"BEFORE: {len(df.index)}")
print(f"DROPPED (gap) : {len(df[(df.gap > 180) & (df.mag < 5.5)].index)}")
df.drop(df[(df.gap > 180) & (df.mag < 5.5)].index, inplace=True)
print(f"DROPPED (dmin): {len(df[(df.dmin >= 10) & (df.mag < 5.5)].index)}")
df.drop(df[(df.dmin >= 10) & (df.mag < 5.5)].index, inplace=True)
print(f"AFTER: {len(df.index)}")

# print(df[((df.gap > 180) | (df.dmin >= 10)) & (df.mag < 5.5)].drop_duplicates().id.count())
# gap > 180 or dmin >= 10 with a mag < 5.5

# df.drop()
```
