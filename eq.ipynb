<h2>Data Cleaning</h2>

<h3>Filtering - </h3>

```python
# Drop duplicates
before=len(df.index)
print(f"BEFORE: {before}")
df.drop_duplicates(inplace=True)
after=len(df.index)
print(f"AFTER: {after}")
print(f"REMOVED: {before-after}")
```

```python
# Drop rows with mag lower than 4
before=len(df.index)
print(f"BEFORE: {before}")
df.drop(df[df.mag<=4].index, inplace=True)
after=len(df.index)
print(f"AFTER: {after}")
print(f"REMOVED: {before-after}")
```

```python
print(f"BEFORE: {len(df.index)}")
print(f"DROPPED (gap) : {len(df[(df.gap > 180) & (df.mag < 5.5)].index)}")
df.drop(df[(df.gap > 180) & (df.mag < 5.5)].index, inplace=True)
print(f"DROPPED (dmin): {len(df[(df.dmin >= 10) & (df.mag < 5.5)].index)}")
df.drop(df[(df.dmin >= 10) & (df.mag < 5.5)].index, inplace=True)
print(f"AFTER: {len(df.index)}")

# print(df[((df.gap > 180) | (df.dmin >= 10)) & (df.mag < 5.5)].drop_duplicates().id.count())
# gap > 180 or dmin >= 10 with a mag < 5.5

# df.drop()
```

<h2>Imputing - </h2>

<p> Imputing nst null and zero values with 1.<br>
As there has to be a minimum of one station that reported the magnitude. </p>

```python
# Using fillna(1) method to replace NaN (null) values with 1.
# Using replace(0, 1) method to replace any zero values with 1.

df['nst'] = df['nst'].fillna(1).replace(0, 1)
```

<p>Replace NA (null) values in the 'magError' column with 0.</p>

```python
# Using the fillna(0) method to replace NA values in the 'magError' column with 0.

df['magError'].fillna(0, inplace = True)
```

<p>To handle the null values in the 'gap' column, we fill them with the median value. <br>
This approach is neutral and doesn't bias the data.</p>

```python
# Using the fillna() method with inplace=True to replace null values in 'gap' with the median of the gap column.

df['gap'].fillna(df['gap'].median(), inplace=True)
```

```python
distinct_magType = df['magType'].unique()
print('Distinct Values in the magType Column - ', distinct_magType)
print(f'The number of distinct magnitude types in the column magType = {len(distinct_magType)}')
```

<p>
Here we have 23 distinct values in the magType column.<br>
To simplify the dataset for analysis,<br> 
We can create broader categories that group the magnitude types based on their method of calculation or the type of seismic waves they measure.<br><br>
Body Wave Magnitudes:<br>
Category Name: "body wave"<br>
Includes: mb, Mb, mb_lg, mblg<br>

Surface Wave Magnitudes:<br>
Category Name: "surface wave"<br>
Includes: ms, ms_20<br>
    
Moment Magnitudes:<br>
Category Name: "moment"<br>
Includes: mw, mwc, mwb, mwr, mww, mwp, ml, Ml, mlg<br>

Duration Magnitudes:<br>
Category Name: "duration"<br>
Includes: md, Md, m, mh, mc, mlr, mlv, Mi<br>

</p>

```python
# Define the translation dictionary
magType_categories = {
    'body wave': ['mb', 'Mb', 'mb_lg', 'mblg'],
    'surface wave': ['ms', 'ms_20'],
    'moment': ['mw', 'mwc', 'mwb', 'mwr', 'mww', 'mwp', 'ml', 'Ml', 'mlg'], 
    'duration': ['md', 'Md', 'm', 'mh', 'mc', 'mlr', 'mlv', 'Mi']
}
# Function to replace values in a column with the corresponding key from the dictionary
def replace_with_category(magType, magTypeDict):
    for category, values in magTypeDict.items():
        if magType in values:
            return category
    return magType

# Assuming 'df' is your DataFrame and 'magType' is the column you want to update
df['magType'] = df['magType'].apply(lambda magType: replace_with_category(magType, magType_categories))
```

<h3> Analysis: </h3>
<li>
Magnitude and Number of Stations (nst):<br>
There is a moderately strong positive correlation between mag (magnitude) and nst (number of stations reporting the earthquake). This suggests that larger earthquakes tend to be recorded by more seismic stations, which is expected since larger quakes are likely to be detected by a broader network.
<br>
<li>Magnitude and Gap:<br>
mag has a moderate negative correlation with gap. This indicates that as the magnitude of an earthquake increases, the gap (largest azimuthal gap between adjacent stations) tends to decrease. This could be due to larger earthquakes being more likely to have comprehensive coverage by seismic networks.
<br>
<li>Depth and Horizontal Error:<br>
There is a moderate positive correlation between depth and horizontalError. This might suggest that as the depth of an earthquake increases, the uncertainty in the horizontal positioning also increases, possibly due to the increased complexity in accurately locating deeper seismic events.
<br>
<li>Magnitude and Magnitude Error (magError):<br>
The correlation between mag and magError is moderately negative. Higher magnitudes might have a smaller reported error, potentially because they are easier to estimate accurately due to stronger seismic signals.
<br>
<li>Number of Stations (nst) and Gap:<br>
There's a strong negative correlation between nst and gap. More stations contributing to an earthquake report can result in a smaller gap, indicating a higher confidence in the earthquake's reported location.
<br>
<li>Dmin and Horizontal Error:<br>
dmin (the horizontal distance to the nearest station from the epicenter) has a strong positive correlation with horizontalError. The further the nearest station is, the greater the uncertainty in the earthquake's horizontal position, which makes sense because more distant stations would likely lead to less precise location measurements.
<br>
<li>Horizontal Error and Depth Error:<br>
There is a positive correlation between horizontalError and depthError, suggesting that when there's more uncertainty in the horizontal location of an earthquake, there tends to be more uncertainty in the depth measurement as well.
    
</li>
